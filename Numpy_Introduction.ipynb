#Comparison between Core Python and Numpy

**Core Python** mean Python without any special modules, i.e. especially without NumPy.

The advantages of Core Python:

1. High-level number objects: integers, floating point
2. Containers: lists with cheap insertion and append methods, dictionaries with fast lookup

Advantages of using Numpy with Python:

1.  Array oriented computing
2.  Efficiently implemented multi-dimensional arrays
3.  Designed for scientific computation


Many advanced Python libraries, such as Scikit-Learn, Scipy, and Keras, make extensive use of the NumPy library. Therefore, if you plan to pursue a career in data science or machine learning, NumPy is a very good tool to master.



```python
x = [2, 3, 4, 5, 6]
y = [a + 2 for a in x] #using list comprehension
#to add 2 to each element in the list x, need to traverse the entire list and add 2 to each element individually. 
print(y)
```

```python
#with the NumPy library:
import numpy as np

nums = np.array([2, 3, 4, 5, 6])

nums2 = nums + 2

print(nums2)

#it is easy to add a scalar value to each element in the list via NumPy. 
#It is not only readable, but also faster when compared to the previous code.
```
